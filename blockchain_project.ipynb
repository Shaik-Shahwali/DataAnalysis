## B2B cross-border transactions on Blockchain technology in various regions worldwide in 2021, with forecasts from 2022 to 2025

```python
b2b_crossborder_transactions = pd.read_excel("c:/test/Blockchain/sheet15.xlsx")
```

```python
b2b_crossborder_transactions
```

```python
import numpy as np
```

```python
years = ['2025','2024','2023','2022','2021','2020']

North_America = [254.3, 159.6, 100.7, 64.1,39.2,20.6]
Latin_America = [36.2,18.2,9.7,5.0,2.6,1.1]
Europe = [466.5, 337.5, 233.6, 152.1, 81.3, 33.3]
Far_East_N_China = [744.7, 536.3, 371.3, 240.3, 109.3, 45.8]
Rest_of_World = [265.6, 190.2, 127.8, 80.7, 46.8, 21.4]
```

```python
sns.set(rc={"figure.figsize":(15, 10)})
w = 0.1
bar1 = np.arange(len(years))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
bar4 = [i+w for i in bar3]
bar5 = [i+w for i in bar4]
bar6 = [i+w for i in bar5]

plt.bar(bar1, North_America, width=w, label="North America")
plt.bar(bar2, Latin_America, width=w, label="Latin America")
plt.bar(bar3, Europe, width=w, label="Europe" )
plt.bar(bar4, Far_East_N_China, width=w, label="FarEast & China")
plt.bar(bar5, Rest_of_World, width=w, label="Rest of World")

plt.xlabel("Regions")
plt.ylabel("USD in Millions")
plt.title("B2B cross-border transactions on Blockchain technology")
plt.xticks(bar1+2*w, years)
plt.legend()
plt.show()
```

## Observation: B2B cross-border transactions on Blockchain technology has been on a rise
